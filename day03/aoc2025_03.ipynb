```python
lines: list[str] = open('input03').read().strip().split('\n')
lines[0]
```

```python
def solve_rec(line: str, num_digits: int) -> str:
    '''A recursive solution. Find the first max and recurse on substring.'''
    if num_digits == 1: return max(line)
    maxi: str = max(line[:-num_digits + 1])
    return maxi + solve_rec(line[line.index(maxi) + 1:], num_digits - 1)
```

```python
%%time
print('Part 1:', sum(int(solve_rec(x, 2)) for x in lines))
print('Part 2:', sum(int(solve_rec(x, 12)) for x in lines))
```
