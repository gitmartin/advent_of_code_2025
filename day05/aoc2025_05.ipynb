### AoC 2025 Day 5

```python
lines = open('input05').read().strip()
intervals_str, nums_str = lines.split('\n\n')
intervals_str = intervals_str.split('\n')
intervals: list[list[int]] = [list(map(int, x.split('-'))) for x in intervals_str]
nums: list[int] = list(map(int, nums_str.split('\n')))
```

#### Part 1

```python
num_contained = 0
for n in nums:
    for i in range(len(intervals)):
        if intervals[i][0] <= n <= intervals[i][1]:
            num_contained += 1
            break
print('Part 1:', num_contained)
```

#### Part 2

```python
# merge intervals
intervals.sort()
for i in range(1, len(intervals)):
    if intervals[i][0] <= intervals[i-1][1]: # start is before end of previous interval
        intervals[i][0] = intervals[i-1][0] # set start to start of previous interval
        intervals[i][1] = max(intervals[i][1], intervals[i-1][1]) # set end
        intervals[i-1] = [] # this interval was merged
print('Part 2:', sum(x[1] - x[0] + 1 for x in intervals if x != []))
```
