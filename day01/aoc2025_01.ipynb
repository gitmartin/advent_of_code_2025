```python
lines: list = open('input01').read().strip().split('\n')
lines = [int(s.replace('L', '-').replace('R','')) for s in lines]
lines[:5]
```

```python
zeros_part1 = 0
zeros_part2 = 0 
pos = 50
for n in lines:
    if n < 0: pos = (100 - pos) % 100 # flip
    pos += abs(n)
    q, pos = divmod(pos, 100)
    zeros_part2 += q
    if n < 0: pos = (100 - pos) % 100 # flip back
    if pos == 0: zeros_part1 += 1
print('Part 1:', zeros_part1)
print('Part 2:', zeros_part2)
```
