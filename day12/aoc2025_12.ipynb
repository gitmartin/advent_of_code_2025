```python
import numpy as np
import math
lines = open('input12').read().strip().split('\n\n')
```

```python
class Shape:
    def __init__(self, grid: np.ndarray, shape_num: int) -> None:
        self.grid = grid
        self.area = int(grid.sum())
        self.shape_num = shape_num
    def __repr__(self) -> str:
        return str((self.shape_num, str(self.grid)))
    
class Region:
    def __init__(self, sizes: tuple, num_shapes: list) -> None:
        self.sizes = sizes
        self.num_shapes = num_shapes
    def __repr__(self) -> str:
        return str(self.sizes) + str(self.num_shapes)
    
def process_shape_input(s: str) -> Shape:
    li = s.split()
    shape_num = int(li.pop(0).split(':')[0])
    m = np.array([list(map(lambda x: 1 if x == '#' else 0, list(x))) for x in li])
    return Shape(m, shape_num)

def process_region_inputs(s: str) -> Region:
    size_str, num_shapes_str = s.split(':')
    size = tuple(map(int,size_str.split('x')))
    num_shapes = list(map(int, num_shapes_str.split()))
    return Region(size, num_shapes)

shapes_pre = [x for x in lines if 'x' not in x]
regions_pre = [x for x in lines if 'x' in x][0].split('\n') 
shapes: list[Shape] = [process_shape_input(x) for x in shapes_pre]
regions: list[Region] = [process_region_inputs(x) for x in regions_pre]
```

```python
shapes_fit = 0
for r in regions:
    total_required_area = np.array(r.num_shapes).dot(np.array([s.area for s in shapes]))
    available_region_area = math.prod(r.sizes)
    if total_required_area <= available_region_area:
        shapes_fit += 1 
        
print('Part 1:', shapes_fit, f'(out of {len(regions)})')
```
