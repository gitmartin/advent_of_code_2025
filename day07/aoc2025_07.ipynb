```python
from collections import defaultdict
lines = open('input07').readlines()
beam_indeces = {lines[0].index('S'): 1}
beam_indeces
```

```python
part1 = 0
for line in lines[1:]:
    beam_indeces_next = defaultdict(int)
    for idx, vals in beam_indeces.items():
        match line[idx]:
            case '^':
                beam_indeces_next[idx-1] += vals 
                beam_indeces_next[idx+1] += vals 
                part1 += 1
            case '.':
                beam_indeces_next[idx] += vals
    beam_indeces = beam_indeces_next

print('Part 1:', part1)
print('Part 2:', sum(beam_indeces.values()))
```
