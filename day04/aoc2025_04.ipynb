### AoC 2025 Day 4

```python
grid_raw = open('input04').read().strip().split('\n')
grid: list[list[str]] = [list(x) for x in grid_raw]
len(grid), len(grid[0])
```

#### Part 1

```python
def count_neighbors(grid: list[list[str]], r: int, c: int) -> int:
    '''Given a row,col (r,c) position in the grid, return number of roll (@) neighbors.'''
    cnt = 0
    for dr in [-1,0,1]:
        for dc in [-1,0,1]:
            if dr == 0 and dc == 0:
                continue
            rr = r + dr
            cc = c + dc
            if 0 <= rr < len(grid) and 0 <= cc < len(grid[0]):
                if grid[rr][cc] == '@':
                    cnt += 1
    return cnt
```

```python
roll_cnt = 0
for r in range(len(grid)):
    for c in range(len(grid[0])):
        if grid[r][c] == '@' and count_neighbors(grid,r,c) < 4:
            roll_cnt += 1
print('Part 1:', roll_cnt)
```

#### Part 2

```python
total_change = 0
while True:
    change = 0 
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] == '@' and count_neighbors(grid, r, c) < 4:
                grid[r][c] = '.'
                change += 1
    total_change += change
    if change == 0:
        break
print('Part 2:', total_change)
```
