```python
# input data preprocessing
ranges_str = open('input02').read().strip().split(',')
ranges: list[tuple] = [tuple(map(int, s.split('-'))) for s in ranges_str]
ranges[:5]
```

```python
import re
part1 = 0
part2 = 0
for lo, hi in ranges:
    for n in range(lo, hi + 1):
        if re.fullmatch(r"(.+)\1", str(n)):  part1 += n
        if re.fullmatch(r"(.+)\1+", str(n)): part2 += n        
            
print("Part 1:", part1)
print("Part 2:", part2)
```
