### AoC 2025 Day 6

```python
import math 
lines = open('input06').readlines()
lines = [x.strip('\n') for x in lines]
[x[:70] for x in lines] # first 70 columns
```

#### Part 1

```python
def do_math(nums: list[str], op: str) -> int:
    match op:
        case '*': return math.prod(int(x) for x in nums)
        case '+': return sum(int(x) for x in nums)
        case _: raise 
        
matrix = [x.split() for x in lines]
num_cols = len(matrix[0])
print('number of columns:', num_cols)
print('first column:', [x[0] for x in matrix])
```

```python
part1 = 0 
for i in range(num_cols):
    part1 += do_math(nums=[x[i] for x in matrix[:-1]], op=matrix[-1][i])
print('Part 1:', part1)
```

#### Part 2

```python
part2 = 0
stack: list[str] = []
for i in range(len(lines[0]) - 1, -1, -1):
    nums = [x[i] for x in lines]
    if all(c == ' ' for c in nums): continue
    stack.append(''.join(nums[:-1]))
    if nums[-1] in '+*':
        part2 += do_math(stack, nums[-1])
        stack = []
print('Part 2:', part2)
```
